```python
while True:
    # Entrada de dados
    numero1 = float(input('Digite o primeiro número: '))
    operacao = input('Digite a operação (+, -, *, /): ')
    numero2 = float(input('Digite o segundo número: '))

    # Processamento e Saída de dados
    if operacao == '+':
        print("Resultado da Soma:", numero1 + numero2)

    elif operacao == '-':
        print("Resultado da Subtração:", numero1 - numero2)

    elif operacao == '*':
        print("Resultado da Multiplicação:", numero1 * numero2)

    elif operacao == '/':
        if numero2 == 0:
            print("Erro: Não é possível dividir por zero!")
        else:
            print("Resultado da Divisão:", numero1 / numero2)

    else:
        print("Operação inválida!")

    # Pergunta se o usuário quer continuar
    continuar = input('\nDeseja fazer outra operação? (s/n): ').lower()

    if continuar != 's':
        print("Programa encerrado. Até logo!")
        break # Quebra o loop e sai do programa






```
